```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# The path is relative to the root of your project
file_path = '../data/../data/sierraleone-bumbuna.csv'
df = pd.read_csv(file_path, parse_dates=['Timestamp'])
```

```python
# See the first 5 rows to confirm it loaded correctly
df.head()
```

```python
# Get a technical summary of the columns (data types, non-null counts)
df.info()
```

```python
# Get summary statistics for all numerical columns
df.describe()
```

```python
# The 'inplace=True' argument modifies the dataframe directly
df.drop(columns=['Comments'], inplace=True)

# Verify the column is gone
df.head()
```

```python
# List of columns to clean
irradiance_cols = ['GHI', 'DNI', 'DHI']

# Loop through each column and set negative values to 0
for col in irradiance_cols:
    df[col] = df[col].apply(lambda x: 0 if x < 0 else x)

# Verify the change by checking the summary statistics again. The 'min' should now be 0.0
df[irradiance_cols].describe()
```

```python
from scipy.stats import zscore

# Calculate Z-scores for the 'GHI' column
df['GHI_zscore'] = zscore(df['GHI'])

# Find all rows where the absolute Z-score is greater than 3
outliers = df[df['GHI_zscore'].abs() > 3]

print(f"Number of outliers detected based on Z-score for GHI: {len(outliers)}")
print("A sample of the detected GHI outliers:")
print(outliers.head())
```

```python
# Remove the temporary Z-score column as it was just for analysis
# We use a try-except block in case the cell is run more than once
try:
    df.drop(columns=['GHI_zscore'], inplace=True)
except KeyError:
    print("GHI_zscore column already removed.")

# Export the cleaned DataFrame to a new CSV file in the 'data' folder
clean_file_path = '../data/sierraleone-clean.csv'
df.to_csv(clean_file_path, index=False)

print(f"Cleaned data for Sierra Leone successfully saved to: {clean_file_path}")
```

```python
# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# Plot 1: Distribution of Ambient Temperature
sns.histplot(df['Tamb'], bins=30, kde=True, ax=axes[0])
axes[0].set_title('Distribution of Ambient Temperature (Tamb)', fontsize=14)
axes[0].set_xlabel('Temperature (C)', fontsize=12)
axes[0].set_ylabel('Frequency', fontsize=12)

# Plot 2: Distribution of Wind Speed
sns.histplot(df['WS'], bins=30, kde=True, ax=axes[1])
axes[1].set_title('Distribution of Wind Speed (WS)', fontsize=14)
axes[1].set_xlabel('Wind Speed (m/s)', fontsize=12)

# Show the plots
plt.tight_layout()
plt.show()
```

```python
# Select a subset of interesting columns for the heatmap
correlation_cols = ['GHI', 'DNI', 'DHI', 'Tamb', 'RH', 'WS']
correlation_matrix = df[correlation_cols].corr()

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Key Weather Variables', fontsize=16)
plt.show()
```
